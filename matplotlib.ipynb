```python
n = np.linspace(1, 50, 50)

BigO = { 'O(1)': np.ones_like(n), 
        'O(log n)': np.log(n), 
        'O(n)': n,
        'O(n log n)': n * np.log(n),
        'O(n^2)': n**2,
        'O(2^n)': 2**n }

plt.figure(figsize=(9, 6))
for key, value in BigO.items():
    plt.plot(n, BigO[key], label=key)

plt.ylim(0, 200)
plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('Big O Time Complexities')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

```python

from matplotlib import cm

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10, 8))
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

# Heatmap

##### Correlation lies between -1 and 1
corr = 1 i.e. positively linearly related

```python
Maths, Science = np.random.randint(12,100, 100), np.random.randint(34,78, 100)
data = pd.DataFrame({'Maths':Maths, 
                     'Science':Science, 
                     'Sst':np.random.randint(30,87, 100),
                     'English':np.random.randint(70,100, 100)
                    })
```

```python
corrmatrix = data.corr()
```

```python
heatmap = plt.imshow(corrmatrix, vmin=-1, vmax=1)
plt.title('Heatmap')
plt.xticks([0,1,2,3],['Maths','Science', 'Social', 'English'])
plt.yticks([0,1,2,3],['Maths','Science', 'Social', 'English'])
plt.colorbar(heatmap); plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [60, 30, 10]  # Percentages
labels = ['Product Development', 'Marketing', 'Operations']
explode = (0.1, 0.1, 0.1)  # Equal explosion for all pieces

# Create figure and axis
plt.figure(figsize=(10, 8))

# Create pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=['white', 'black', 'lightgray'], # Using only black and white colors autopct='%1.0f%%', # Show percentages shadow=True, startangle=90, textprops={'color': 'grey', 'fontsize': 14})

# Add a title
plt.title('Funding Allocation', pad=20, size=14)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
plt.show()
# Save the plot
plt.savefig('funding_allocation.png', bbox_inches='tight', dpi=300)
plt.close()
```
